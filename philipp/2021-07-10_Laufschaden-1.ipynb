```python
# kidaq 725rpm 0%
```

```python
mdf_obj_725rmp0Prozent = MDF(folder_kidaq / "725rpm" / "725rpm@0%.mf4")
df_obj_725rmp0Prozent = mdf_obj_725rmp0Prozent.to_dataframe()
df_obj_725rmp0Prozent[["p1", "p2"]].rolling(window=100000, step=100000).median().plot(title="725rpm@0%", grid=True, ylim=(-0.008, 0.008)).get_figure().savefig("725rpm@0%.png")
```
